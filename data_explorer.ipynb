```python
import pandas as pd
```

```python
def normalize_dataframe(df):
    data_list = []

    for column in df.columns:
        if type(df[column][0]) == dict:
            data_list.append(pd.json_normalize(df[column]))
        else:
            data_list.append(df[column])    

    normalized_df = pd.concat(data_list, axis=1)

    return normalized_df
```

```python
def print_dataframe_basic_info(df):

    print('\n')
    print('======================================================================')
    print('===================== CHECKING COLUMNS DATA TYPE =====================')
    print('======================================================================')
    print('\n')

    df.info()

    print('\n')
    print('======================================================================')
    print('============= SEARCHING FOR UNIQUE VALUES IN EACH COLUMN =============')
    print('======================================================================')
    print('\n')

    for column in df.columns:

        print(f'Column: {column}, Unique values: {df[column].unique()}')


```

```python
file = 'data/Telco-Customer-Churn.json'
raw_df = pd.read_json(file)
raw_df
```

```python
normalized_df = normalize_dataframe(raw_df)
normalized_df
```
